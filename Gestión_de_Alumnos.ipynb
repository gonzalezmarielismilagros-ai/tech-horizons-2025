```python
estudiantes=[]
def agregar_estudiante(lista):
    print("ESTUDIANTE NUEVO")
    nombre=input("Nombre del estudiante: ").strip()
    while True:
        try:
            edad=int(input("Edad: "))
            promedio=float(input("Promedio o Nota Final: "))
            if 0<=edad and 0.0<=promedio<=10.0:
                break
            else:
                print("Revisa la edad (no negativa) y el promedio (0.0-10.0).")
        except ValueError:
            print("Edad debe ser entero y promedio debe ser n√∫mero")
    nuevo_estudiante={"nombre":nombre,"edad":edad,"promedio":promedio}
    lista.append(nuevo_estudiante)
    print(f"‚úÖ ¬°{nombre} ({edad} a√±os) a√±adido al sistema!")
def mostrar_estudiantes(lista):
    if not lista:
        print("El sistema est√° vac√≠o, a√±ade estudiantes")
        return
    print("LISTA COMPLETA DE ESTUDIANTES")
    for i,estudiante in enumerate(lista):
        print(f"{i+1}. Nombre: {estudiante['nombre']:<15} | Edad: {estudiante['edad']:<3} Promedio: {estudiante['promedio']:.2f}")
    print("-" * 50)
def buscar_estudiante(lista):
    if not lista:
        print("El sistema est√° vac√≠o, no hay a quien buscar.")
        return
    nombre_buscado=input("Dime el nombre a buscar: ").strip().lower()
    encontrado=False
    print("\n--- RESULTADOS DE B√öSQUEDA ---")
    for estudiante in lista:
        if nombre_buscado in estudiante['nombre'].lower():
            print(f"‚úÖ Encontrado: {estudiante['nombre']}")
            print(f"   Edad: {estudiante['edad']}")
            print(f"   Promedio: {estudiante['promedio']:.2f}")
            encontrado=True
    if not encontrado:
        print(f"‚ùå No se encontr√≥ nadie llamado '{nombre_buscado}'.")
def calcular_estadisticas(lista):
    if not lista:
        print("No hay datos para estad√≠sticas.")
        return
    total_estudiantes=len(lista)
    suma_edades=0
    suma_promedios=0.0
    for estudiante in lista:
        suma_edades+=estudiante['edad']
        suma_promedios+=estudiante['promedio']
    promedio_edad=suma_edades/total_estudiantes
    promedio_general=suma_promedios/total_estudiantes
    print("ESTAD√çSTICAS DEL GRUPO")
    print(f"Total de estudiantes: {total_estudiantes}")
    print(f"Promedio de Edad del Grupo: {promedio_edad:.1f} a√±os")
    print(f"Promedio General de Notas: {promedio_general:.2f}")
if __name__=="__main__":
    while True:
        print("\n"+"="*50)
        print("üè´ SISTEMA DE ESTUDIANTES COMPLETO")
        print("="*50)
        print("1. Agregar Estudiante")
        print("2. Mostrar Todos los Estudiantes")
        print("3. Buscar Estudiante por Nombre")
        print("4. Calcular Estad√≠sticas del Grupo")
        print("5. Salir")
        opcion=input("\nElige una opci√≥n (1-5): ")
        print("-"*50)
        if opcion=='1':
            agregar_estudiante(estudiantes)
        elif opcion=='2':
            mostrar_estudiantes(estudiantes)
        elif opcion=='3':
            buscar_estudiante(estudiantes)
        elif opcion=='4':
            calcular_estadisticas(estudiantes)
        elif opcion=='5':
            print("Sistema apagado üëã")
            break
        else:
            print("Tienes que usar del 1 al 5")
```

```python
estudiantes=[
    {"nombre":"Andrea","edad":16,"promedio":9.5,"grado":11,"materias_favoritas":["Matem√°ticas","F√≠sica"]},
    {"nombre":"Braulio","edad":15,"promedio":8.2,"grado":10,"materias_favoritas":["Historia","Arte"]},
    {"nombre":"Carla","edad":17,"promedio":7.0,"grado":11,"materias_favoritas":["Qu√≠mica","Biolog√≠a"]},
    {"nombre":"Daniel","edad":15,"promedio":9.8,"grado":10,"materias_favoritas":["Programaci√≥n","Ingl√©s"]},
    {"nombre":"Elena","edad":16,"promedio":8.9,"grado":11,"materias_favoritas":["Literatura","Filosof√≠a"]}]
def agregar_estudiante(lista):
    print("\nESTUDIANTE NUEVO")
    nombre=input("Nombre del estudiante: ").strip()
    while True:
        try:
            edad=int(input("Edad:"))
            promedio=float(input("Promedio o Nota Final:"))
            grado=int(input("Grado/Curso:"))
            materias_str=input("Materias favoritas:")
            materias_favoritas=[m.strip() for m in materias_str.split(',')]
            if 0<=edad and 0.0<=promedio<=10.0 and grado>0:
                break
            else:
                print("Revisa los datos.")
        except ValueError:
            print("Edad/Grado deben ser enteros y Promedio debe ser n√∫mero.")
    nuevo_estudiante={"nombre":nombre,"edad":edad,"promedio":promedio,"grado":grado,"materias_favoritas":materias_favoritas}
    lista.append(nuevo_estudiante)
    print(f"‚úÖ {nombre} a√±adido al sistema")
def mostrar_estudiantes(lista):
    if not lista:
        print("A√±ade estudiantes")
        return
    print("\nLISTA COMPLETA DE ESTUDIANTES")
    for i,estudiante in enumerate(lista):
        materias_str=", ".join(estudiante.get('materias_favoritas',[]))
        print(f"{i+1}. Nombre: {estudiante['nombre']:<12} | Edad: {estudiante['edad']:<2} | Grado: {estudiante.get('grado','N/A'):<3} | Promedio: {estudiante['promedio']:.2f} | Materias Fav: {materias_str}")
    print("-" * 100)
def buscar_estudiante(lista):
    if not lista:
        print("El sistema est√° vac√≠o, no hay a quien buscar.")
        return
    nombre_buscado=input("Dime el nombre a buscar: ").strip().lower()
    encontrado=False
    print("\n--- RESULTADOS DE B√öSQUEDA ---")
    for estudiante in lista:
        if nombre_buscado in estudiante['nombre'].lower():
            print(f"‚úÖ Encontrado: {estudiante['nombre']}")
            print(f"   Edad: {estudiante['edad']} | Grado: {estudiante.get('grado','N/A')}")
            print(f"   Promedio: {estudiante['promedio']:.2f} | Materias: {', '.join(estudiante.get('materias_favoritas',[]))}")
            encontrado=True
    if not encontrado:
        print(f"‚ùå No se encontr√≥ nadie llamado '{nombre_buscado}'.")
def eliminar_estudiante(lista):
    if not lista:
        print("No hay nadie que eliminar.")
        return
    mostrar_estudiantes(lista)
    while True:
        try:
            indice=int(input("Escribe el N√öMERO del estudiante a eliminar: "))-1
            if 0<=indice<len(lista):
                estudiante_removido=lista.pop(indice)
                print(f"üóëÔ∏è ¬°{estudiante_removido['nombre']} eliminado del sistema!")
                break
            else:
                print("Ese n√∫mero no est√° en la lista.")
        except ValueError:
            print("Tiene que ser un n√∫mero entero.")
def actualizar_estudiante(lista):
    if not lista:
        print("No hay nadie para actualizar.")
        return
    mostrar_estudiantes(lista)
    while True:
        try:
            indice=int(input("Escribe el N√öMERO del estudiante a actualizar: "))-1
            if 0<=indice<len(lista):
                estudiante=lista[indice]
                print(f"\nActualizando a: {estudiante['nombre']}")
                nuevo_promedio=float(input(f"Nuevo promedio (actual: {estudiante['promedio']}): "))
                nuevo_grado=int(input(f"Nuevo grado (actual: {estudiante['grado']}): "))
                estudiante['promedio']=nuevo_promedio
                estudiante['grado']=nuevo_grado
                print(f"‚úÖ {estudiante['nombre']} actualizado. Nuevo promedio: {nuevo_promedio:.2f}")
                break
            else:
                print("Ese n√∫mero no est√° en la lista.")
        except ValueError:
            print("Aseg√∫rate de meter n√∫meros correctos.")
def calcular_estadisticas(lista):
    if not lista:
        print("No hay datos para estad√≠sticas.")
        return
    total_estudiantes=len(lista)
    suma_edades=sum(e['edad'] for e in lista)
    suma_promedios=sum(e['promedio'] for e in lista)
    promedio_edad=suma_edades/total_estudiantes
    promedio_general=suma_promedios/total_estudiantes
    print("\n--- ESTAD√çSTICAS DEL GRUPO ---")
    print(f"Total de estudiantes: {total_estudiantes}")
    print(f"Promedio de Edad del Grupo: {promedio_edad:.1f} a√±os")
    print(f"Promedio General de Notas: {promedio_general:.2f}")
if __name__=="__main__":
    while True:
        print("\n"+"="*50)
        print("SISTEMA DE ESTUDIANTES SUPER COMPLETO")
        print("="*50)
        print("1. Agregar nuevo estudiante")
        print("2. Mostrar todos los estudiantes")
        print("3. Buscar Estudiante por Nombre")
        print("4. Eliminar Estudiante")
        print("5. Actualizar informaci√≥n (Promedio/Grado)")
        print("6. Calcular estad√≠sticas del grupo")
        print("7. Salir")
        opcion=input("\nElige una opci√≥n (1-7): ")
        print("-"*50)
        if opcion=='1':
            agregar_estudiante(estudiantes)
        elif opcion=='2':
            mostrar_estudiantes(estudiantes)
        elif opcion=='3':
            buscar_estudiante(estudiantes)
        elif opcion=='4':
            eliminar_estudiante(estudiantes)
        elif opcion=='5':
            actualizar_estudiante(estudiantes)
        elif opcion=='6':
            calcular_estadisticas(estudiantes)
        elif opcion=='7':
            print("Vuelve pronto üëã")
            break
        else:
            print("Tienes que usar del 1 al 7")
```
